```python
datas = [
    {'thread': 1, 'time': 40, 'usage': 10},
    {'thread': 2, 'time': 21, 'usage': 18},
    {'thread': 3, 'time': 15, 'usage': 25},
    {'thread': 4, 'time': 12, 'usage': 30},
    {'thread': 6, 'time': 8.3, 'usage': 45},
    {'thread': 8, 'time': 6.6, 'usage': 55},
    {'thread': 12, 'time': 5.1, 'usage': 80},
    {'thread': 16, 'time': 4.3, 'usage': 100},
]
```

```python
import matplotlib.pyplot as plt

# Extract data into separate lists
threads = [d['thread'] for d in datas]
times = [d['time'] for d in datas]
usages = [d['usage'] for d in datas]

# Create figure and axis
fig, ax = plt.subplots()

# Plot time vs thread
ax.plot(threads, times, label='Time')
ax.set_xlabel('Thread')
ax.set_ylabel('Time')

# Plot usage vs thread
ax2 = ax.twinx()
ax2.plot(threads, usages, label='Usage', color='red')
ax2.set_ylabel('Usage (%)')

# Set title and legend
ax.set_title('Time and Usage per Thread')
ax.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show plot
plt.show()
```
